# MATPLOTLIB

```python
# BarPlot-Average Ticket Price by Airline
airline_price = airline.groupby('airline')['price'].mean().sort_values(ascending=False)

plt.figure(figsize=(8, 6))
airline_price.plot(kind='bar', color='skyblue')
plt.title('Average Ticket Price by Airline')
plt.xlabel('Airline')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
#histogram-Ticket Price Distribution
plt.figure(figsize=(8, 5))
plt.hist(airline['price'], bins=50, color='teal', edgecolor='black')
plt.title('Distribution of Ticket Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#scatterplot-Price vs Days Left
plt.figure(figsize=(8, 5))
plt.scatter(airline['days_left'], airline['price'], color='darkgreen')
plt.title('Ticket Price vs Days Left to Departure')
plt.xlabel('Days Left')
plt.ylabel('Price')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#boxplot-Ticket Price by Class
classes = airline["class"].unique()
price_data = [airline[airline["class"] == c]["price"] for c in classes]

plt.figure(figsize=(8, 6))
plt.boxplot(price_data, tick_labels=classes)
plt.title('Ticket Price Distribution by Class')
plt.xlabel('Class')
plt.ylabel('Price')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#piechart-Flights by Departure Time
dep_time_counts = airline['departure_time'].value_counts()

plt.figure(figsize=(6, 6))
dep_time_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=plt.cm.Pastel1.colors)
plt.title('Flights by Departure Time Category')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
#lineplot-Average Economy Ticket Price vs Days Left
eco_flights = airline[airline["class"] == "Economy"]
avg_price_by_day = eco_flights.groupby("days_left")["price"].mean()


plt.figure(figsize=(8, 5))
plt.plot(avg_price_by_day.index, avg_price_by_day.values, color="green", marker="o")
plt.title("Average Economy Ticket Price vs Days Left", fontsize=14)
plt.xlabel("Days Left to Departure")
plt.ylabel("Average Ticket Price (INR)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Seaborn

```python
#scatterplot - Price vs Duration

sns.scatterplot(data=airline, x="duration", y="price", hue="class")
plt.title("Price vs Flight Duration by Class")
plt.xlabel("Flight Duration (hours)")
plt.ylabel("Price (INR)")
plt.show()
```

```python
#Box Plot — Price Distribution by Airline
plt.figure(figsize=(8, 6))
sns.boxplot(data=airline, x="airline", y="price")
plt.xticks(rotation=45)
plt.title("Ticket Price Distribution by Airline")
plt.xlabel("Airline")
plt.ylabel("Price (INR)")
plt.tight_layout()
plt.show()
```

```python
#Violin Plot — Price vs Class

sns.violinplot(data=airline, x="class", y="price", hue="class" , palette="Set2")
plt.title("Price Distribution by Ticket Class")
plt.xlabel("Class")
plt.ylabel("Price (INR)")
plt.show()
```

```python
#Count Plot — Number of Flights by City

plt.figure(figsize=(8, 5))
sns.countplot(data=airline, x="source_city", order=airline["source_city"].value_counts().index)
plt.title("Number of Flights from Each City")
plt.xlabel("Source City")
plt.ylabel("Flight Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Distribution Plot — Ticket Price
sns.histplot(data=airline, x="price", kde=True, bins=40, color="skyblue")
plt.title("Distribution of Ticket Prices")
plt.xlabel("Price (INR)")
plt.ylabel("Number of Flights")
plt.show()
```

```python
import streamlit as st
import pandas as pd

def load_data():
    df = pd.read_csv('airlines_flights_data.csv')  
    return df

df = load_data()

#app title
st.title("Flight Price EDA Dashboard")

# Show raw data
if st.checkbox("Show Raw Data"):
    st.write(df.head())

# Sidebar filters
st.sidebar.header("Filter Options")
airline_filter = st.sidebar.multiselect("Select Airline", options=df['airline'].unique(), default=df['airline'].unique())
class_filter = st.sidebar.multiselect("Select Class", options=df['class'].unique(), default=df['class'].unique())

# Apply filters
filtered_df = df[(df['airline'].isin(airline_filter)) & (df['class'].isin(class_filter))]

# Display some key info
st.subheader("Filtered Data Summary")
st.write(f"Number of flights: {filtered_df.shape[0]}")

# Plot: Price distribution
st.subheader("Price Distribution")
fig, ax = plt.subplots()
sns.histplot(filtered_df['price'], kde=True, ax=ax)
st.pyplot(fig)

# Boxplot: Airline vs Price
st.subheader("Price by Airline")
fig2, ax2 = plt.subplots(figsize=(12,6))
sns.boxplot(x='airline', y='price', data=filtered_df, ax=ax2)
plt.xticks(rotation=90)
st.pyplot(fig2)

# Lineplot: Days left vs price
st.subheader("Days Left vs Price")
fig3, ax3 = plt.subplots()
sns.lineplot(x='days_left', y='price', data=filtered_df, ax=ax3)
st.pyplot(fig3)
```
